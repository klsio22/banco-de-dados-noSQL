```python
import requests

# Substitua 'demo' pela sua própria chave de API da Alpha Vantage
API_KEY = 'sua_chave_aqui'  # Obtenha uma em https://www.alphavantage.co/support/#api-key

# URL para a consulta (usando TIME_SERIES_INTRADAY como exemplo)
symbol = 'IBM'  # Substitua pelo símbolo da ação desejada
interval = '5min'  # Intervalo de tempo entre os dados (1min, 5min, etc.)
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={symbol}&interval={interval}&apikey={API_KEY}'

# Fazendo a requisição
response = requests.get(url)

# Convertendo a resposta para JSON
data = response.json()

# Verificando o conteúdo
print(data)
```

```python
import pandas as pd

# Extraindo a série temporal
time_series = data.get('Time Series (5min)', {})

# Convertendo para DataFrame
df = pd.DataFrame.from_dict(time_series, orient='index')

# Renomeando colunas para facilitar
df.columns = ['open', 'high', 'low', 'close', 'volume']

# Convertendo os índices (datas) para o formato datetime
df.index = pd.to_datetime(df.index)

# Ordenando os dados pela data
df = df.sort_index()

# Convertendo valores para float (exceto volume)
df[['open', 'high', 'low', 'close']] = df[['open', 'high', 'low', 'close']].astype(float)

# Exibindo os dados processados
print(df.head())
```
