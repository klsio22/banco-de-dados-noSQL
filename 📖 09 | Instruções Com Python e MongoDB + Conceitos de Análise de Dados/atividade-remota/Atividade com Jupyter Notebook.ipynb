```python
import pandas as pd

# Carregar o arquivo CSV
df = pd.read_csv('titanic.csv')

# Converter o DataFrame para JSON
data_json = df.to_dict(orient='records')
```

```python
import pymongo

# Conectar ao MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["titanic_db"]
collection = db["passengers"]
```

```python
collection.insert_many(data_json)
print("Dados inseridos com sucesso!")
```

```python
def inserir_registro(dados):
    collection.insert_one(dados)
    print("Registro inserido com sucesso!")
```

```python
def exibir_todos():
    registros = list(collection.find())
    return pd.DataFrame(registros)
```

```python
def exibir_com_filtro(filtro):
    registros = list(collection.find(filtro))
    return pd.DataFrame(registros)
```

```python
def deletar_registro(filtro):
    collection.delete_one(filtro)
    print("Registro deletado com sucesso!")
```

```python
def remover_registro_por_nome(nomes):
    for nome in nomes:
        resultado = collection.delete_one({"Name": nome})
        if resultado.deleted_count > 0:
            print(f'Registro com o nome "{nome}" foi removido com sucesso.')
        else:
            print(f'Registro com o nome "{nome}" não encontrado.')


nomes_para_remover = ["John Doe", "João", "João Silva"]

remover_registro_por_nome(nomes_para_remover)
```
